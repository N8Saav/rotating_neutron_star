```python
import numpy as np
import matplotlib.pyplot as plt

kepler_rotating_GM1L_h = np.loadtxt('GM1L_h_rotating.dat', unpack=True, skiprows=2)
static_GM1L_h = np.loadtxt('GM1L_h_static.dat', unpack=True, skiprows=2)

A2_0_rotating_mass_density_GM1L_h = np.loadtxt('A2_0_rotating_mass_density_GM1L_h.dat', unpack=True, skiprows=2)
A1_8_rotating_mass_density_GM1L_h = np.loadtxt('A1_8_rotating_mass_density_GM1L_h.dat', unpack=True, skiprows=2)
A1_6_rotating_mass_density_GM1L_h = np.loadtxt('A1_6_rotating_mass_density_GM1L_h.dat', unpack=True, skiprows=2)

plt.plot(static_GM1L_h[1], static_GM1L_h[3])
plt.plot(kepler_rotating_GM1L_h[1], kepler_rotating_GM1L_h[3])

plt.plot(A2_0_rotating_mass_density_GM1L_h[1], A2_0_rotating_mass_density_GM1L_h[3])
plt.plot(A1_8_rotating_mass_density_GM1L_h[1], A1_8_rotating_mass_density_GM1L_h[3])
plt.plot(A1_6_rotating_mass_density_GM1L_h[1], A1_6_rotating_mass_density_GM1L_h[3])

plt.xlabel('Central Energy Density [MeV fm$^{-3}$]')
plt.ylabel('Gravitational Mass [Solar Masses]')
plt.legend(["Static","Kepler Freq",'Baryon Mass = 2.0 [Solar Masses]','Baryon Mass = 1.8','Baryon Mass = 1.6'])
plt.show()
```
