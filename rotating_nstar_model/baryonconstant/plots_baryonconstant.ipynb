```python
import numpy as np
import matplotlib.pyplot as plt

kepler_rotating_GM1L_h = np.loadtxt('GM1L_h_rotating.dat', unpack=True, skiprows=2)
kepler_static_GM1L_h = np.loadtxt('GM1L_h_static.dat', unpack=True, skiprows=2)
static_mass_density_GM1L_h = np.loadtxt('static_mass_density_A1-8_GM1L_h.dat', unpack=True, skiprows=2)
rotating_mass_density_GM1L_h = np.loadtxt('rotating_mass_density_A1-8_GM1L_h.dat', unpack=True, skiprows=2)
rotating_frequency_GM1L_h = np.loadtxt('rotating_frequency_A1-8_GM1L_h.dat', unpack=True, skiprows=3)

plt.plot(rotating_mass_density_GM1L_h[1], rotating_mass_density_GM1L_h[3])
plt.plot(kepler_static_GM1L_h[1], kepler_static_GM1L_h[3])
plt.plot(kepler_rotating_GM1L_h[1], kepler_rotating_GM1L_h[3])
plt.xlabel('Central Energy Density [MeV fm$^{-3}$]')
plt.ylabel('Gravitational Mass [Solar Masses]')
plt.show()
```
